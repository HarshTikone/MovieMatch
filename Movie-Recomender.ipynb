```python
def convert(obj):
  list_1 = []
  for i in ast.literal_eval(obj):
    list_1.append(i['name'])

  return list_1
```

```python
df['genres'] = df['genres'].apply(convert)
```

```python
df['keywords'] = df['keywords'].apply(convert)
```

```python
df['production_companies'] = df['production_companies'].apply(convert)
```

```python
def convertcast(obj):
  list_1 = []
  count = 0
  for i in ast.literal_eval(obj):
    if count < 5 :
        list_1.append(i['name'])

  return list_1
```

```python
df['cast'] = df['cast'].apply(convertcast)
```

```python
def convertcrew(obj):
  list_1 = []
  for i in ast.literal_eval(obj):
    if i['job'] == "Director":
      list_1.append(i['name'])
      break
  return list_1
```

```python
df['crew'] = df['crew'].apply(convertcrew)
```

```python
import sklearn
from sklearn.feature_extraction.text import CountVectorizer
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
PS = PorterStemmer()
```

```python
def porstem(text):
  y = []

  for i in text.split():
    y.append(PS.stem(i))
  
  return " ".join(y)
```

```python
df["tags"] = df["tags"].apply(porstem)
```

```python
df.head(3)
```

```python
df['tags'][1]
```

```python
cv = CountVectorizer(max_features = 5000 , stop_words = 'english')
vec = cv.fit_transform(df['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similar_movies = cosine_similarity(vec)
```

```python
sorted(list(enumerate(similar_movies[0])),reverse = True, key = lambda x : x[1])[1:10]
```

```python
def recomd(movie):
  mov_ind = df[df['title']==movie].index[0]
  dis = similar_movies[mov_ind]
  mov_list = sorted(list(enumerate(dis)), reverse = True , key = lambda x: x[1])[1:10]

  for i in mov_list :
    print(df.iloc[i[0]].title)
```

```python
recomd("Avatar")
```
